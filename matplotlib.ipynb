angles

```python

import numpy as np
y=np.random.normal([35,25,25,15])
mylabels=(['apple','banna','chery','dates'])

plt.pie(y,labels=mylabels,startangle=90)
plt.show()
```

```python
#explode
import numpy as np
y=np.random.normal([35,25,25,15])
mylabels=(['apple','banna','chery','dates'])
myexplode=[0.2,0,0,0]
plt.pie(y,labels=mylabels,explode=myexplode,shadow=True)
plt.show()
```

Colors

```python
#colors --r red hexadecimal all are valid
import numpy as np
y=np.random.normal([35,25,25,15])
mylabels=(['apple','banna','chery','dates'])
mycolors = ['black', 'hotpink', 'b', '#4CAF50']

plt.pie(y,labels=mylabels,colors = mycolors)
plt.show()
```

```python
import numpy as np
y=np.random.normal([35,25,25,15])
mylabels=(['apple','banna','chery','dates'])
mycolors = ['black', 'hotpink', 'b', '#4CAF50']
plt.pie(y,labels=mylabels)
plt.legend()
plt.show()
```

```python
import numpy as np
y=np.random.normal([35,25,25,15])
mylabels=(['apple','banna','chery','dates'])
mycolors = ['black', 'hotpink', 'b', '#4CAF50']
plt.pie(y,labels=mylabels)
plt.legend(title="fruits")
plt.show()
```
